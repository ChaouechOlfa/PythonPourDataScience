```python
import matplotlib.pyplot as plt
x=range(1,10)
y=[4,3,2,1,0,1,2,3,4]
```

```python
#Affichage de graphique
plt.plot(x,y)
```

```python
#Diagramme Ã  barres
plt.bar(x,y)
```

```python
#Si on souhaite des barres horizontales
plt.barh(x,y)
```

```python
#Diagramme Circulaire
p=[10,1,3,4,2]
plt.pie(p)
```

```python
Etiquettes= 'Produit1','Produit2','Â¨Produit3','Produit4','Produit5'
p=[10,1,3,4,2]
plt.pie(p,labels=Etiquettes)
plt.show()
```

```python
#Affichage de % de chaque produit
plt.pie(p,labels=Etiquettes,autopct='%1.1f%%')
plt.show()
```

```python
e=(0,0,0,0,0.1)
plt.pie(p,explode=e,labels=Etiquettes,autopct='%1.1f%%')
plt.show()
plt.savefig('fig1.png')
```

```python
x = np.linspace(0, 10, 300)
rng = np.random.RandomState(0)
y = np.cumsum(rng.randn(300, 3), 0)

plt.plot(x, y[:, 0], label='Curve 1')
plt.plot(x, y[:, 1], label='Curve 2')
plt.plot(x, y[:, 2], label='Curve 3')

plt.legend()
plt.show()
```

```python
#Utilisant seaborn:
import seaborn as sns
sns.set()
plt.plot(x,y)
plt.legend('ABC',ncol=2,loc='upper right')
```
