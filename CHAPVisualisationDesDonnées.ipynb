```python
x = np.linspace(0, 10, 300)
rng = np.random.RandomState(0)
y = np.cumsum(rng.randn(300, 3), 0)

plt.plot(x, y[:, 0], label='Curve 1')
plt.plot(x, y[:, 1], label='Curve 2')
plt.plot(x, y[:, 2], label='Curve 3')

plt.legend()
plt.show()
```
